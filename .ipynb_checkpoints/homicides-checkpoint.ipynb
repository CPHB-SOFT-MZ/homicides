```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

url = "https://www.dropbox.com/sh/4b3dc4rgbrvy2ia/AABC51KK9XLyrTAVvm02Qb0na/database.csv?dl=1"
response = requests.get(url, params={'downloadformat' : 'csv'})

# print(response.headers)

fname = response.headers['Content-Disposition'].split('=')[1].split(";")[0].replace('\"', "")

if response.ok:
    with open(fname, 'wb') as f:
        f.write(response.content)

csv = pd.read_csv(fname, dtype={"Perpetrator Age": object})

dd = csv.as_matrix()
```

```python
victim_ethnicity, count = np.unique(dd[:,14], return_counts=True)
print(victim_ethnicity[np.argmax(count)])

df = pd.DataFrame(data=count, index=victim_ethnicity, columns=["Homicides"])
df.plot.bar()
plt.show()
```

```python
weapons, count = np.unique(dd[(dd[:,15] == "Male")][:,20], return_counts=True)
print(weapons[np.argmax(count)])
df3 = pd.DataFrame(data=count, index=weapons, columns=["Homicides by"])
df3.plot.pie(subplots=True)
plt.show()
```

```python
weapons, count = np.unique(dd[(dd[:,15] == "Female")][:,20], return_counts=True)
print(weapons[np.argmax(count)])
df3 = pd.DataFrame(data=count, index=weapons, columns=['Homicides'])
df3.plot.bar()
plt.show()
```

```python
victim_ages = np.unique(dd[:,12])
print("Lowest age of a victim:", np.argmin(victim_ages))
print("Highest age of a victim:", np.argmax(victim_ages))
```

```python
ages = np.array(dd[:,12], dtype=np.float)
print(np.average(ages))
```

```python
_, male_count = np.unique(dd[(dd[:,15] == "Male")][:,15], return_counts=True)
_, female_count = np.unique(dd[(dd[:,15] == "Female")][:,15], return_counts=True)

print("For every female there are approximately", int(np.round(male_count[0] / female_count[0])), "males commiting a homicide")

```

```python
states, count = np.unique(dd[:,5], return_counts=True)
df4 = pd.DataFrame(data=count, index=states, columns=['Homicides']).sort_values(by="Homicides", ascending=False).head(10)
df4.plot.bar()
plt.show()
```
